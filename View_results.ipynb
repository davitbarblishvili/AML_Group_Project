```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("results.csv"); df
```

```python
df_avg = df.groupby("Model").mean(); df_avg
```

```python
def model_score(acc, f1, ft, pt):
    return (acc*100+f1*100)/np.log(ft*1000+2*pt*1000)

df_avg['score'] = model_score(df_avg['Accuracy'],df_avg['F1-Score'], df_avg['Fit-Time'], df_avg['Predict-Time'])
df_avg
```

```python
df_avg = df_avg.sort_values(by='score', axis=0, ascending=False); df_avg
```

```python
df_avg.to_csv("results_avg.csv")
```
